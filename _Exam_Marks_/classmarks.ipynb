```python
import pandas as pd
import matplotlib.pyplot as plt
dt=pd.read_csv("class_marks.csv")
```

```python
dt.plot()
plt.show()
```

```python
dt.columns
```

```python
dt['Q1'] = dt['Q1aM4'] + dt['Q1bM6']
plt.hist(dt['Q1'], color='red', edgecolor='black', bins=9)
plt.xlabel("Marks scored in Q1")
plt.ylabel("No. of students")
plt.title("Histogram of Q1")
plt.show()
```

# The distribution is moderately spread with more students scoring in the middle range.

```python
dt['Q2'] = dt['Q2aM6'] + dt['Q2bM4']
plt.hist(dt["Q2"], color='yellow', edgecolor='black', bins=10)
plt.xlabel("Marks scored in Q2")
plt.ylabel("No. of students")
plt.title("Histogram of Q2")
plt.show()
```

# The distribution is relatively even with a slight right skew, indicating more students scored lower.


```python
dt['Q4'] = dt['Q4aM3'] + dt["Q4bM7"]
plt.hist(dt["Q4"], color='green', edgecolor='black', bins=10)
plt.xlabel("Marks scored in Q4")
plt.ylabel("No. of students")
plt.title("Histogram of Q4")
plt.show()
```

#  The data shows a left-skewed distribution with most students scoring in the lower range.

```python
plt.hist(dt["Q5M10"], color='purple', edgecolor='black', bins=10)
plt.xlabel("Marks scored in Q5")
plt.ylabel("No. of students")
plt.title("Histogram of Q5")
plt.show()
```

# The scores are concentrated in the higher range, with a few students achieving significantly higher marks.

```python
plt.hist(dt['Total'], color='brown', edgecolor='black', bins=10)
plt.xlabel("Marks scored in Q8")
plt.ylabel("No. of students")
plt.title("Histogram of Q8")
plt.show()
```
