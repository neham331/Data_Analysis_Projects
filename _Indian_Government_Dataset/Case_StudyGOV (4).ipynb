### District-wise Tele-Law Case Registration and Advice Enabled Data from FY 2021-22 to 2024-25

# Dataset Description


* This dataset provides district-wise data on Tele-Law case registrations and advice-enabled services across Indian states and Union Territories (UTs) from FY 2021-22 to 2024-25. The Tele-Law initiative aims to provide legal aid to citizens, especially in rural areas, through Common Service Centers (CSCs).



# Key Attributes in the Dataset:
* Category: Specifies whether the data is for cases registered or advice enabled.
* States/UTs: The Indian state or AUT where the data is recorded.
* Districts: The district-wise breakdown of case registration and legal advice.
* No. of CSCs: Number of Common Service Centers (CSCs) in a district facilitating Tele-Law services.
* Gender-wise Cases: Number of cases registered by females and males separately.
* Total Cases Registered: Sum of male and female cases in a district.
* Caste-wise Cases: Breakdown of cases by General, OBC, SC, and ST categories.

# Purpose of the Dataset:
* To analyze state-wise and district-wise case registrations and trends.
* To identify which states/districts have the highest case registrations.
* To study the gender and caste-wise distribution of cases.
* To evaluate the role of Common Service Centers (CSCs) in providing legal aid.
* Key Observations from the Dataset:
* Highest Cases Registered: Uttar Pradesh has the highest number of cases registered.
* Top District in Telangana: Adilabad has the most cases among Telangana districts.
* Gender-Wise Distribution: More males (61%) than females (39%) have used Tele-Law services.
* Caste-Wise Distribution: OBCs (44%) have the highest registered cases, followed by SCs (28%).
* Role of CSCs: Districts with a higher number of CSCs tend to have more cases registered.

# Filtering Out Total Values from 'States/UT's' Column

```python
df_filtered = df[~df["States/UT's"].str.contains("Total", case=False, na=False)]
print(df_filtered.head())
```

```python
df["States/UT's"] = df["States/UT's"].replace(
    {"Dadra And Nagar Haveli & Daman & Diu": "Dadra and Nagar Haveli & Daman & Diu"})
df = df[df["States/UT's"] != "Total (36 States/UT's)"]
statewise_total = df.groupby("States/UT's")["Total"].sum().sort_values(ascending=False).head(10)
print(statewise_total)
```

```python

missing_values = df.isnull().sum()
print("Missing Values in Each Column:\n", missing_values)
```

```python
# Fill missing values for numeric columns with their mean
numeric_cols = ['No. of CSCs', 'Female', 'Male', 'Total', 'General', 'OBC', 'SC', 'ST', 'Total.1']
df[numeric_cols] = df[numeric_cols].apply(lambda x: x.fillna(x.mean()))

# Verify missing values after filling
print(df.isnull().sum())
```

```python
df.drop(columns=["Total.1"], inplace=True)
```

# Caste Distribution Anomalies

```python
caste_totals = df[["General", "OBC", "SC", "ST"]].sum()
caste_percentage = (caste_totals / caste_totals.sum()) * 100
print("National-Level Caste Distribution:\n", caste_percentage)
```

```python
district_caste_percentage = df.groupby("Districts")[["General", "OBC", "SC", "ST"]].sum()
district_caste_percentage = district_caste_percentage.div(district_caste_percentage.sum(axis=1), axis=0) * 100
outlier_districts = district_caste_percentage[(district_caste_percentage > (caste_percentage + 10)).any(axis=1)]
print("Districts with Caste Distribution Anomalies:\n", outlier_districts)
```

# High-Risk Districts (Cases vs. State Average)

```python
state_avg_cases = df.groupby("States/UT's")["Total"].mean()
df["Cases vs. State Avg"] = df.apply(lambda x: x["Total"] / state_avg_cases[x["States/UT's"]], axis=1)
high_risk_districts = df[df["Cases vs. State Avg"] > 2] 
print("High-Risk Districts (Cases > 2x State Avg):\n", high_risk_districts[["States/UT's", "Districts", "Cases vs. State Avg"]].head(10))
```

# Outlier Detection Using Z-Score

```python
from scipy.stats import zscore
df["Z-Score"] = zscore(df["Total"])
outliers = df[(df["Z-Score"] > 3) | (df["Z-Score"] < -3)]  
print("Outlier Districts (Z-Score > 3 or < -3):\n", outliers[["States/UT's", "Districts", "Total", "Z-Score"]])
```
